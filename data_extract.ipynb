```python
target_tests = [ 'baseline',
    'visual_nback_1', 'visual_nback_2', 'visual_nback_3',
    'auditory_nback_1','auditory_nback_2', 'auditory_nback_3',
    'dual_nback_1', 'dual_nback_2', 'dual_nback_3'
]
```

```python
def baseline_correct(df, exclude_columns=['class', 'subject']):
    # Identify columns to correct (all columns except excluded ones)
    columns_to_correct = [col for col in df.columns if col not in exclude_columns]
    
    # For each subject
    corrected_dfs = df.copy()
    for feature in columns_to_correct:
        baseline_vale = df.loc[df['class'] == 'baseline',feature]
        corrected_dfs[feature] = (df[feature] - baseline_vale.values[0]) / baseline_vale.values[0]
        
        
    # Combine back into one DataFrame
    return corrected_dfs
```

```python
# Initialize an empty list to store DataFrames
df_subjects = pd.DataFrame()

# Loop through all files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.csv'):
        # Construct full file path
        file_path = os.path.join(folder_path, filename)

        # Read CSV file into a DataFrame
        df = pd.read_csv(file_path)
        subject = filename.split('_')[0]
        df['subject'] = int(subject)

        # Apply baseline correction
        df_bc = baseline_correct(df)
        
        df_subjects = pd.concat([df_subjects,df_bc])

        # Add DataFrame to list
        #print(f"Read {filename} with {len(df)} rows")

print(f"Combined DataFrame has {len(df_subjects)} rows")
```

```python
# Filter only the tests we're interested in
df_subjects = df_subjects[df_subjects['class'].isin(target_tests)]
# Get rid of NaN
df_subjects.fillna(0, inplace=True)
#df_subjects[df_subjects['class'] == 'baseline'].head(50)
df_subjects.head(40)
```

```python
# Melt ratings data to long format
rating_cols = ['baseline','visual_nback_1', 'visual_nback_2', 'visual_nback_3',
    'auditory_nback_1','auditory_nback_2', 'auditory_nback_3',
    'dual_nback_1', 'dual_nback_2', 'dual_nback_3']

df_ratings_long = df_ratings.melt(
    id_vars=['subject', 'sex', 'age'],
    value_vars=rating_cols,
    var_name='test',
    value_name='rating'
)

# Merge with physiological data
df_merged = pd.merge(
    df_subjects,
    df_ratings_long,
    left_on=['subject', 'class'],
    right_on=['subject', 'test'],
    how='inner'
)

# Drop unnecessary columns
df_merged = df_merged.drop(columns=['test'])

# Step 1: Merge and Prepare Data (same as before)
X = df_merged.drop(columns=['subject', 'class', 'rating', 'sex', 'age'])
y = df_merged['rating']  
groups = df_merged['subject']

df_merged
```

```python
df_merged = df_merged.rename(columns={'class': 'test_phase'})
df_merged = df_merged.rename(columns={'rating': 'label'})

df_merged = df_merged.drop(columns=['pnn50', 'pnn50_max'])

# Save to csv
df_merged.to_csv('df_merged.csv', index=False)
# Save to pickle
df_merged.to_pickle('df_merged.pkl')
```
